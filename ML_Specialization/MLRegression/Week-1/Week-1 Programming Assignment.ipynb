```python
import graphlab
```

```python
sales = graphlab.SFrame('kc_house_data.gl')
```

```python
train_data,test_data = sales.random_split(0.8,seed=0)
```

```python
sales
```

```python
def simple_linear_regression(input_feature,output):
    sigmaxi = input_feature.sum()
    print sigmaxi
    sigmayi = output.sum()
    print sigmayi
    sigmaxiyi = (input_feature*output).sum()
    print sigmaxiyi
    sigmaxisq = (input_feature*input_feature).sum()
    print sigmaxisq
    N = len(input_feature)
    print N
    w1 = (N*sigmaxiyi-sigmaxi*sigmayi)/(N*sigmaxisq-sigmaxi*sigmaxi)
    w0 = output.mean()-w1*(input_feature.mean())
    print "Intercept: "+str(w0)
    print "Slope: "+str(w1)
    return (w0,w1)
```

```python
test_feature = graphlab.SArray(range(5))
test_out = graphlab.SArray(1+1*test_feature)
(test_intercept,test_slope) = simple_linear_regression(test_feature,test_out)
print "Intercept: "+str(test_intercept)
print "Slope: "+str(test_slope)
```

```python
sqft_intercept,sqft_slope = simple_linear_regression(train_data['sqft_living'],train_data['price'])
```

```python
def get_regression_predictions(input_feature,w0,w1):
    return w0+w1*input_feature
```

```python
print "%.2f" %(get_regression_predictions(2650,sqft_intercept,sqft_slope))
```

```python
def RSS(input_feature,output,w0,w1):
    pred_y = w0+w1*input_feature
    temp = (output-pred_y)
    temp = temp*temp
    return temp.sum()
```

```python
RSS(test_feature,test_out,test_intercept,test_slope)
```

```python
print "Value of RSS is "+str(RSS(train_data['sqft_living'],train_data['price'],sqft_intercept,sqft_slope))
```
